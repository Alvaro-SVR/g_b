```python
import scqubits as scq
import numpy as np
import matplotlib.pyplot as plt
```

## Transmon

![transmon](img/transmon.webp)

```python
transmon_param_list = [
    {"EJ": 20.0, "EC": 20.0},
    {"EJ": 20.0, "EC": 4.0},
    {"EJ": 50.0, "EC": 1.0}
]

transmons = [
    scq.Transmon(EJ=p["EJ"], EC=p["EC"], ng=0.0, ncut=20)
    for p in transmon_param_list
]

ng_vals = np.linspace(-1.0, 1.0, 200)

for t in transmons:
    t.plot_evals_vs_paramvals('ng', ng_vals, evals_count=4)
```

## Fluxonium

![fluxonium](img/fluxonium.webp)

```python
fluxonium_param_list = [
    {"EJ": 200.0, "EC": 1000.0, "EL": 4.0},
    {"EJ": 200.0, "EC": 100.0,  "EL": 4.0},
    {"EJ": 200.0, "EC": 10.0,   "EL": 4.0},
    {"EJ": 200.0, "EC": 1.0,    "EL": 4.0},
    {"EJ": 200.0, "EC": 40.0,   "EL": 100.0},
    {"EJ": 200.0, "EC": 10.0,   "EL": 100.0},
]

fluxoniums = [
    scq.Fluxonium(EJ=p["EJ"], EC=p["EC"], EL=p["EL"], flux=0.0, cutoff=50)
    for p in fluxonium_param_list
]

flux_vals = np.linspace(-1.0, 1.0, 200)

for f in fluxoniums:
    f.plot_evals_vs_paramvals('flux', flux_vals, evals_count=5)
```

## Tunable Transmon

![tunable_transmon](img/tunable_transmon.webp)

```python
tunabletransmon_param_list = [
    {"EJ": 20.0, "EC": 20.0},
    {"EJ": 20.0, "EC": 4.0},
    {"EJ": 50.0, "EC": 1.0}
]

tunabletransmons = [
    scq.TunableTransmon(EJmax=p["EJ"], d=0.1, EC=p["EC"], flux=0.0, ng=0.0, ncut=50)
    for p in tunabletransmon_param_list
]

flux_vals = np.linspace(-1.0, 1.0, 100)

for t in tunabletransmons:
    t.plot_evals_vs_paramvals('flux', flux_vals, evals_count=5)
```

## FluxQubit

![flux_qubit](img/fluxqubit.webp)

```python
fluxqubits_param_list = [
    {"EJ": 20.0, "alpha": 0.6, "ECg": 40.0}
]

fluxqubits = [
    scq.FluxQubit(
        EJ1=p["EJ"], EJ2=p["EJ"], EJ3=p["alpha"] * p["EJ"],
        ECJ1=1.0, ECJ2=1.0, ECJ3=1.0 / p["alpha"],
        ECg1=p["ECg"], ECg2=p["ECg"],
        ng1=0.0, ng2=0.0,
        flux=0.5,
        ncut=50
    )
    for p in fluxqubits_param_list
]

flux_vals = np.linspace(-1.0, 1.0, 50)

for t in fluxqubits:
    t.plot_evals_vs_paramvals('flux', flux_vals, evals_count=5)
```

## ZeroPi

![zeropi](img/zeropi.webp)

```python
phi_grid = scq.Grid1d(-6 * np.pi, 6 * np.pi, 200)

zeropi_param_list = [
    {"EJ": 0.25, "EL": 1e-2,  "ECJ": 0.5,  "ECS": 1e-3, "ng": 0.1, "flux": 0.23},
    {"EJ": 0.5,  "EL": 5e-3,  "ECJ": 1.0,  "ECS": 2e-3, "ng": 0.2, "flux": 0.25},
    {"EJ": 1.0,  "EL": 1e-3,  "ECJ": 2.0,  "ECS": 1e-4, "ng": 0.0, "flux": 0.5},
    {"EJ": 0.8,  "EL": 2e-2,  "ECJ": 0.5,  "ECS": 5e-3, "ng": 0.3, "flux": 0.1},
    {"EJ": 0.3,  "EL": 5e-3,  "ECJ": 0.8,  "ECS": 2e-3, "ng": 0.15,"flux": 0.4}
]

zero_pis = [
    scq.ZeroPi(
        grid=phi_grid,
        EJ=p["EJ"],
        EL=p["EL"],
        ECJ=p["ECJ"],
        EC=None,
        ECS=p["ECS"],
        ng=p["ng"],
        flux=p["flux"],
        ncut=10
    )
    for p in zeropi_param_list
]

flux_vals = np.linspace(-1.0, 1.0, 100)

for zp in zero_pis:
    zp.plot_evals_vs_paramvals('flux', flux_vals, evals_count=4)
```

```python
phi_grid = scq.Grid1d(-6 * np.pi, 6 * np.pi, 200)

fullzeropi_param_list = [
    {"EJ": 0.25, "EL": 1e-2,  "ECJ": 0.5,  "EC": 0.001, "dEJ": 0.05, "dCJ": 0.05, "dEL": 0.01, "dC": 0.08, "ng": 0.1, "flux": 0.23},
    {"EJ": 0.5,  "EL": 5e-3,  "ECJ": 1.0,  "EC": 0.002, "dEJ": 0.10, "dCJ": 0.05, "dEL": 0.02, "dC": 0.10, "ng": 0.2, "flux": 0.25},
    {"EJ": 1.0,  "EL": 1e-3,  "ECJ": 2.0,  "EC": 0.001, "dEJ": 0.02, "dCJ": 0.10, "dEL": 0.01, "dC": 0.05, "ng": 0.0, "flux": 0.5},
    {"EJ": 0.8,  "EL": 2e-2,  "ECJ": 0.5,  "EC": 0.003, "dEJ": 0.08, "dCJ": 0.05, "dEL": 0.02, "dC": 0.07, "ng": 0.3, "flux": 0.1},
    {"EJ": 0.3,  "EL": 5e-3,  "ECJ": 0.8,  "EC": 0.002, "dEJ": 0.05, "dCJ": 0.08, "dEL": 0.01, "dC": 0.09, "ng": 0.15,"flux": 0.4}
]

full_zero_pis = [
    scq.FullZeroPi(
        grid=phi_grid,
        EJ=p["EJ"],
        EL=p["EL"],
        ECJ=p["ECJ"],
        EC=p["EC"],
        dEJ=p["dEJ"],
        dCJ=p["dCJ"],
        dEL=p["dEL"],
        dC=p["dC"],
        ng=p["ng"],
        flux=p["flux"],
        ncut=10,
        zeropi_cutoff=10,
        zeta_cutoff=30
    )
    for p in fullzeropi_param_list
]

flux_vals = np.linspace(-1.0, 1.0, 100)

for fzp in full_zero_pis:
    fzp.plot_evals_vs_paramvals('flux', flux_vals, evals_count=4)
```

## Variaciones

### Symmetric Transmon

![symmetric_transmon](img/symmetric_transmon.jpeg)

```python
tunable_transmon = scq.TunableTransmon(
    EJmax=15.0,
    d=0.0,
    EC=0.3,
    flux=0.0,
    ng=0.0,
    ncut=50
)

flux_vals = np.linspace(-1.0, 1.0, 200)

E0, E1, E2 = [], [], []

for flux in flux_vals:
    tunable_transmon.flux = flux
    evals = tunable_transmon.eigenvals(evals_count=3)
    E0.append(evals[0])
    E1.append(evals[1])
    E2.append(evals[2])

E0, E1, E2 = np.array(E0), np.array(E1), np.array(E2)

omega_01 = E1 - E0
omega_12 = E2 - E1

plt.figure(figsize=(8,5))
plt.plot(flux_vals, omega_01, label=r'$\omega_{01}$', lw=2)
plt.plot(flux_vals, omega_12, label=r'$\omega_{12}$', lw=2)
plt.xlabel("Magnetic flux bias")
plt.ylabel("Frequency (GHz)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Asymmetric Transmon

![asymmetric_transmon](img/asymmetric_transmon.jpeg)

```python
gamma = 2.5
d = (gamma - 1) / (gamma + 1)

tunable_transmon = scq.TunableTransmon(
    EJmax=15.0,
    d=d,
    EC=0.3,
    flux=0.0,
    ng=0.0,
    ncut=50
)

flux_vals = np.linspace(-1.0, 1.0, 200)

E0, E1, E2 = [], [], []

for flux in flux_vals:
    tunable_transmon.flux = flux
    evals = tunable_transmon.eigenvals(evals_count=3)
    E0.append(evals[0])
    E1.append(evals[1])
    E2.append(evals[2])

E0, E1, E2 = np.array(E0), np.array(E1), np.array(E2)
omega_01 = E1 - E0
omega_12 = E2 - E1

plt.figure(figsize=(8,5))
plt.plot(flux_vals, omega_01, label=r'$\omega_{01}$', lw=2)
plt.plot(flux_vals, omega_12, label=r'$\omega_{12}$', lw=2)
plt.xlabel("Magnetic flux bias")
plt.ylabel("Frequency (GHz)")
plt.title(fr"$\gamma={gamma}$")
plt.ylim((0.0, 8.0))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
