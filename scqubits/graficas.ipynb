```python
import scqubits as scq
import numpy as np
```

## Transmon

```python
param_list = [
    {"EJ": 20.0, "EC": 20.0},
    {"EJ": 20.0, "EC": 4.0},
    {"EJ": 50.0, "EC": 1.0}
]

transmons = [
    scq.Transmon(EJ=p["EJ"], EC=p["EC"], ng=0.0, ncut=20)
    for p in param_list
]

ng_vals = np.linspace(-1.0, 1.0, 200)

for t in transmons:
    t.plot_evals_vs_paramvals('ng', ng_vals, evals_count=4)
```

## Fluxonium

```python
param_list = [
    {"EJ": 200.0, "EC": 1000.0, "EL": 4.0},
    {"EJ": 200.0, "EC": 100.0,  "EL": 4.0},
    {"EJ": 200.0, "EC": 10.0,   "EL": 4.0},
    {"EJ": 200.0, "EC": 1.0,    "EL": 4.0},
    {"EJ": 200.0, "EC": 40.0,   "EL": 100.0},
    {"EJ": 200.0, "EC": 10.0,   "EL": 100.0},
]

fluxoniums = [
    scq.Fluxonium(EJ=p["EJ"], EC=p["EC"], EL=p["EL"], flux=0.0, cutoff=50)
    for p in param_list
]

flux_vals = np.linspace(-1.0, 1.0, 200)

for f in fluxoniums:
    f.plot_evals_vs_paramvals('flux', flux_vals, evals_count=5)
```
